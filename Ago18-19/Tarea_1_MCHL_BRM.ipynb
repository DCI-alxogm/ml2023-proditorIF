# **Regresión lineal por mínimos cuadrados**
Es un método común para estimar los coeficientes de las ecuaciones de regresión lineal que describen la relación entre una mas variables independientes cuantitativas y una variable dependiente. Según el número de variables, podemos hacer una regresión simple o múltiple. Se llama regresión de los mínimos cuadrados por qué se basa en la minimización del error cuadrático.
En este caso nos intersa el caso de la regresión simple es decir una sola variable dependiente.
Se propone el modelo.


---


$\hat{y_i}=mx_i+b $\
LA funcion error seria la siguiente:\
$E(x_i,m,b)=\frac{1}{N}\sum {[y_i-(mx_i+b)^2]}$\
Buscamos minimizar la función error por medio de los parametros b y m.
Esto se logra mediante las condiciones siguientes:\
$\frac{∂E}{\partial m}= 0\quad$   $\frac{∂E}{\partial b}=0$\
$\frac{∂^2E}{\partial m^2}>0\quad$   $\frac{∂^2E}{\partial b^2}>0$\
Despues de un procedimiento matematico empleando la expresion del error y las condiciones dadas se llega a las expresiones para los parametros.\

m=$\frac{N\sum(x_iy_i)-\sum x_i \sum y_i}{N\sum x_i^2-(\sum x_i)^2}$\
b = $\bar{y}-m\bar{x}$









```python
#Importamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
#Definimos las funciones de los parametros
def m(x,y):
  N=len(x)
  XY=sum(x*y)
  X = sum(x)
  Y=sum(y)
  X2=sum(x**2)
  M= (N*XY-X*Y)/(N*X2-X**2)
  return M
def b(x,y,m):
   N=len(x)
   X = sum(x)
   Y=sum(y)
   B=(Y-m*X)/N
   return B
#Extraemos los datos del archivo txt
drive.mount('/content/drive')
file ="/content/drive/MyDrive/Machine_Learn/olympic100m_men.txt"
data=np.loadtxt(file,delimiter=",").T
#Asigamos los datos
x=data[0]
y=data[1]
#Graficamos los datos sin ajustar
plt.xlabel("Fecha")
plt.ylabel("Velocidad")
plt.title("Tiempos record/Sin ajuste")
plt.scatter(x,y)
#Se realiza el ajuste
M =m(x,y)
B =b(x,y,M)
Y= M*x+B
print(M)
```
