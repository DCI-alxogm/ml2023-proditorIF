```python
from numpy.linalg import linalg
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
file ="/content/drive/MyDrive/Machine_Learn/olympic100m_men.txt"
data=np.loadtxt(file,delimiter=",").T
#Asigamos los datos
x=data[0]
t=data[1]
def Wpoly(x, y, n):
     X=np.vstack([np.ones_like(x)]) #matriz
     for i in range(1,n+1):
      Xa=np.vstack([x**i])
      X=np.concatenate([X, Xa])
      Xm=np.matmul(X,X.T) #X^T*X
      invx =np.linalg.inv(Xm)
      Xt =np.matmul(X,y) #X^T*t
      W=np.matmul(invx,Xt)#Se calcula W
     return W
# Se normalizan los datos
x = (x - 1896) / 4
plt.plot(x, t,'o',label="Datos")
Y=0
n=8 # Se define el grado maximo
for i in range(1,n+1):
    W=Wpoly(x[:17], t[:17],i) #Se toma un conjunto de entrenamineto de 17 elementos
    Y=0
    for i in range(len(W)):
        Y+= W[i] * x**i
    plt.plot(x, Y, label=f'Orden {i}')

plt.legend()
plt.xlabel('Fecha')
plt.ylabel('Tiempo (s)')
plt.title('Ajuste por minimos cuadrados hasta orden 8')
plt.ylim(0, 25)
plt.xlim(0, 27)



```
