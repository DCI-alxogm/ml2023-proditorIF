### Introducción
La expansión del universo fue descubierta por primera vez en lo estudios de supernovas tipo IA posteriormente confirmadas por datos de (CMB) y oscilaciones acústicas bariónicas.\
Actualmente existen múltiples modelos que buscan explicar esta expansión acelerada del universo, el modelo más conocido actualmente es el modelo $\Lambda$CBM el cual propone que parte de la energía existente en el universo se presenta en algo conocido como “Energía Oscura” la cual ejerce una presión negativa responsable de dicha expansión.\
En este proyecto se busca aplicar los conceptos de inferencia Bayesiana, mediante el método de Metropolis-Hastings para estimar parámetros presentes en el modelo  ΛCBM
###Ecuación de Friedmann
Partimos de la ecuación de Einstein\
$G_{\mu\nu}=8\pi G T_{\mu\nu}$\
La ecuación de Einstein describe como la distribución de energía del universo curva el espacio tiempo y a su vez como esta curvatura del espacio tiempo cambia la distribución de energía del universo.\
Del lado izquierdo de la ecuación se encuentra la información de la curvatura y del lado derecho se encuentra lo referente a la energía y momento contenidos en el universo.\
Las ecuaciones de Friedmann son una consecuencia directa de la aplicación de la ecuación de campo de Einstein a las métricas de Friedmann-Lemaître-Robertson-Walker (FLRW). Estas métricas son una descripción matemática de la geometría del espacio-tiempo en un universo homogéneo e isotrópico, que es una suposición fundamental en la cosmología a gran escala.\
$(\frac{̇\dot{a}}{a})^2=\frac{8\pi G}{3}\rho-\frac{k}{a^2}$\
El factor de escala se utiliza para describir cómo los objetos (como galaxias) se alejan entre sí debido a la expansión del universo. A medida que el factor de escala aumenta, el universo se expande, y a medida que disminuye, el universo se contrae.
###Componentes
El modelo cosmológico estándar ΛCDM establece que los principales componentes del universo son   materia, radiación, energía oscura y la posibilidad de curvatura .

**Materia:** La materia se puede dividir en dos categorías la materia bariónica y la materia oscura. La materia bariónica, que incluye átomos y partículas subatómicas, es la forma de materia que forma estrellas, galaxias y la mayoría de las estructuras visibles en el cosmos. Por otro lado la materia oscura representa la mayor parte de materia contenida en el universo se sabe que la única interacción de este tipo de materia con la materia “Ordinaria” es de tipo gravitacional, además la materia oscura no interacciona con el campo electromagnético ya que no absorbe ni emite radiación electromagnética.

**Radiación:** La radiación en cosmología se refiere a partículas con velocidades relativistas, donde la densidad de energía está dominada por el término cinético. En este contexto, dos ejemplos destacados de partículas de radiación son los fotones y los neutrinos. Aunque los neutrinos tienen una masa extremadamente pequeña, viajan a velocidades relativistas, lo que significa que se mueven a una fracción significativa de la velocidad de la luz. En términos cosmológicos, tanto los fotones como los neutrinos se consideran un tipo de radiación debido a su comportamiento cinético y a su contribución a la densidad de energía en el universo en expansión.
### Parámetros
Los parámetros cosmológicos son cantidades fundamentales que se utilizan para caracterizar y describir las propiedades del universo a gran escala. Su objetivo principal es proporcionar una descripción cuantitativa de los fenómenos astrofísicos y cosmológicos observables en el universo actual.
Los parámetros más comunes son el parámetro de densidad, el parámetro de Hubble y el parámetro de desaceleración.\
**Parámetro de Hubble:** Representa la tasa actual de expansión del universo. Es esencial para calcular la edad del universo y predecir cómo se expandirá en el futuro.\
Se denota por H y esta defindio por la siguiente expresión: $H=\frac{\dot a}{a}$\
**Parámetro de densidad:**Este parámetro describe la composición del universo en términos de sus componentes principales siendo materia radiación y energía oscura.\
Este parámetro describe la densidad relativa de diferentes componentes del universo en relación con la densidad critica, siendo la densidad critica la densidad media que debe tener el universo para que su expansión sea exactamente igual a 0.\

El parámetro de densidad esta dado por la siguiente expresión:
$\Omega=\frac{\rho}{\rho_c}$
Donde la es la densidad total que contiene las contribuciones de cada componente (materia, radiación y energía oscura ).\
$$\rho=\rho_m+\rho_r+\rho_{Λ}+\rho_k$$\
Donde la representa una densidad hipotética para la curvatura.
La densidad critica esta dada por : $\rho_c=\frac{3H_0^2}{8πG}$\
Dónde $H_0$ es el parámetro de Hubble en la actualidad.
Como ya se mencionó se tendrá un parámetro de densidad para componente principal del universo y además se agrega un parámetro de densidad para la curvatura teniendo así $\Omega_k$, $\Omega_m$, $\Omega_r$ y $\Omega_{Λ}$
De forma que un compoente particular se calcula de la sigueinte manera:\
$$\Omega_k=\frac{8πG}{3H_0^2}\rho_k$$
**Parámetro de Desaceleración:** Este parámetro cuantifica la velocidad de la expansión del universo y si esa expansión se está acelerando o desacelerando debido a la influencia de la materia y la energía oscura.
### Modelo $ΛCDM$
El modelo  $\Lambda\text{CDM}$  (Lambda Cold Dark Matter) es el modelo cosmológico estándar que describe la evolución y estructura del Universo a gran escala. Este modelo se basa en los siguientes conceptos fundamentales:

1. **$(\Lambda)$**: Representa la constante cosmológica, que se refiere a una forma de energía oscura uniforme que llena todo el espacio y tiene efectos gravitacionales repulsivos. En otras palabras, es una energía oscura que genera una fuerza de expansión en el Universo, contrarrestando la fuerza gravitatoria atractiva de la materia.

2. **CDM**: Se refiere a la materia oscura fría, que es una forma de materia no luminosa que no interactúa con la luz ni con otras formas de materia electromagnética. La materia oscura fría es esencial para explicar la formación de estructuras a gran escala en el Universo, como galaxias y cúmulos de galaxias.

El modelo $\Lambda\text{CDM}$ se basa en la relatividad general de Einstein y propone que el unievrso que está compuesto principalmente por:

- **Materia Oscura Fría**: Esta es la forma principal de materia en el Universo y tiene efectos gravitatorios atractivos. Constituye la mayor parte de la densidad de materia en el Universo.

- **Energía Oscura $(\Omega_\Lambda)$**: La constante cosmológica (\(\Lambda\)) representa la energía oscura. Aunque no se comprende completamente, se cree que es responsable de la expansión acelerada del Universo.

- **Materia Bariónica $(\Omega_b)$**: Esta es la materia normal, como átomos, que constituye estrellas, planetas y galaxias. Aunque es una fracción pequeña de la densidad de materia total, es fundamental para la formación de estructuras a pequeña escala.

Además de estos componentes principales, el modelo también considera la radiación $\Omega_r$ en el Universo temprano, que incluye la radiación cósmica de fondo de microondas (CMB) y otras formas de radiación.

El modelo $\Lambda\text{CDM}$ se basa en observaciones y mediciones detalladas, como el CMB, las observaciones de supernovas, las mediciones de lentes gravitacionales y la distribución de galaxias a gran escala. Es capaz de explicar una amplia variedad de observaciones cosmológicas y proporciona una descripción coherente de la evolución del Universo.

En resumen este modelo plantea a la energía oscura como la responsable de la expansión acelerada del universo  y interpreta esta energía oscura como la manifestación efectiva de una constante cosmológica denotada por $Λ$ , esta constante se puede añadir a las ecuaciones de campo de Einstein.
$$G_{\mu\nu}-Λg_{\mu\nu}=8\pi G T_{\mu\nu}$$
Con estas nuevas ecuaciones de campo es posible derivar la ecuación de Friedman que considere el efecto de esta constante.
$$H^2=\frac{8\pi G}{3}\rho+\frac{Λ}3{}-\frac{k}{a^2}$$

#### Apliación del método
Se estimaran los parámetros del modelo empleando al algoritmo de metrópolis.\
Partimos de le ecuación de Friedmann con el corrimiento al rojo  H(z)

$$H^2=H_0^2(\frac{\Omega_m}{a^3}+\frac{\Omega_r}{a^4}+\frac{\Omega_k}{a^2}+\Omega_{Λ})$$

Donde el valor actual  del parámetro de Hubble esta dado por:
$$H_0= 100\frac{km}{sMpc}h$$
Debemos tomar en cuenta que restringiremos el modelo a uno con radiación despreciable, es decir:
$$H^2=H_0^2(\frac{\Omega_m}{a^3}+\frac{\Omega_k}{a^2}+\Omega_{Λ})$$
Y dónde a representa el factor de escala del universo el cual se relaciona con el corrimiento al rojo mediante la siguiente expresión
$$a=\frac{1}{1+z}$$
De froma que la ecuación de Friedmann para el modelo que se empleara sera:
$$H^2 = H_0^2(\Omega_m(1+z)^3+\Omega_{Λ}+\Omega_k(1+z)^2)$$
 Se estimarán los valores de $\Omega_{Λ}$,$\Omega_{m}$ y $H_0$
 a partir de una serie de valores medidos de z y su correspondiente valor de H calculado con ese Z, tambien se incluye un conjunto de deatos que  contiene las incertidumbres asociadas a las mediciones de H0 en los mismos puntos de z. Estas incertidumbres representan la dispersión o error en las mediciones de H0 a lo largo de los diferentes valores de "z"\
 Es decir z_0 [0] es un solo valor de z medidio y con este se calculó un valor de $H_0$ el cual corresponde a H_o[0] y la incertidumbre para este caso sera sigma_o[0]

```python

# Datos observacionales de z y H_0
z_o = np.array([0.07, 0.09, 0.12, 0.17, 0.1791,
                0.1993, 0.2, 0.24, 0.27, 0.28,
                0.30, 0.31, 0.34, 0.35, 0.3519,
                0.36, 0.38, 0.3802, 0.4,
                0.4004, 0.4247, 0.43, 0.44, 0.4497,
                0.47, 0.4783, 0.48, 0.51,
                0.52, 0.56, 0.57, 0.59, 0.5929,
                0.6, 0.61, 0.64, 0.6797, 0.73, 0.75,
                0.7812, 0.8754, 0.88, 0.9, 1.037,
                1.3, 1.363, 1.43, 1.53, 1.75,
                1.965, 2.3, 2.33, 2.34, 2.36])

H_o = np.array([69, 69, 68.6, 83, 75,
                75, 72.9, 79.69, 77, 88.8,
                81.7, 78.18, 83.8, 82.7, 83, 79.94,
                81.5, 83, 82.04, 77, 87.1, 86.45,
                84.81, 92.8, 89, 80.9, 87.79, 90.4, 94.35,
                93.34, 96.8, 98.48, 104, 87.9, 97.3, 98.82,
                92, 97.3, 98.8, 105, 125, 90, 117, 154, 168,
                160, 177, 140, 202, 186.5, 224, 224, 222, 226])

sigma_o = np.array([19.6, 12, 26.2, 8, 4,
                    5, 29.6, 2.99, 14, 36.64,
                    6.22, 4.74, 3.66, 8.4, 14,
                    3.38, 1.9, 13.5, 2.03, 10.2,
                    11.2, 3.97, 1.83, 12.9, 34,
                    9, 2.03, 1.9, 2.64, 2.3,
                    3.4, 3.18, 13, 6.1, 2.1,
                    2.98, 8, 7, 33.6, 12, 17,
                    40, 23, 20, 17, 33.6,
                    18, 14, 40, 50.4, 8, 8, 7, 8])

# Modelo
def lambda_cdm_model(z, H0, omega_m, omega_l):
    omega_k = 1 - omega_m - omega_l
    result = (omega_m * (1 + z)**3 + omega_l + omega_k * (1 + z)**2)*H0**2
    return result

# Función para calcular el logaritmo del likelihood
def calculate_log_likelihood(H0, omega_m, omega_l, z_data, H_data, sigma_data):
    log_likelihood = 0
    for z, H, sigma in zip(z_data, H_data, sigma_data):
        model_H = lambda_cdm_model(z, H0, omega_m, omega_l)
        log_likelihood += -0.5 * ((H**2 - model_H) / (2 * H * sigma))**2
    return log_likelihood

# Función para calcular el logaritmo del prior
def calculate_log_prior(H0, omega_m, omega_l):
    if 50 <= H0 <= 100 and 0 <= omega_m <= 1 and 0 <= omega_l <= 1:
        return 0.0  # Logaritmo del prior plano
    else:
        return -np.inf  # Fuera de los límites, prior cero

# Se estalecen valores inciales y desviaciones
H0 = 65
omega_m = 0.3
omega_l = 0.7
# Se asigan las desviaciones
std_prop = [0.72,0.019,0.033]
# Número de iteraciones y quemado
N = 30000
# Listas para almacenar los pasos de cada parámetro
samples = []
# Valores iniciales
H0_current = H0
omega_m_current =omega_m
omega_l_current = omega_l

# Tasa de aceptación
acceptance_rate = 0

# Algoritmo Metropolis-Hastings
for i in range(N):
     # Propuesta de nuevos valores a partir de una distribución gaussiana
    next_H0 = np.random.normal(H0_current, std_prop[0])
    next_omega_m = np.random.normal(omega_m_current, std_prop[1])
    next_omega_l = np.random.normal(omega_l_current, std_prop[2])
    # Calcula el logaritmo del posterior para los valores actuales y propuestos
    log_prob_current = calculate_log_likelihood(H0_current, omega_m_current, omega_l_current, z_o, H_o, sigma_o) + calculate_log_prior(H0_current, omega_m_current, omega_l_current)
    log_prob_next = calculate_log_likelihood(next_H0, next_omega_m, next_omega_l, z_o, H_o, sigma_o) + calculate_log_prior(next_H0, next_omega_m, next_omega_l)

    r = log_prob_next - log_prob_current
    u = np.log(np.random.rand()) #Parte Aleatoria
    # Decide si aceptar o rechazar la propuesta
    if u < r:
        H0_current = next_H0
        omega_m_current = next_omega_m
        omega_l_current = next_omega_l
        acceptance_rate += 1

    samples.append((H0_current, omega_m_current, omega_l_current))

acceptance_rate /= N

# Resultados
print("Valores medios de H0, omega_m, omega_l:", np.mean(samples, axis=0))
print("Tasa de aceptación:", acceptance_rate)

# Graficar la evolución de la cadena de Markov
result_sampling = np.array(samples)
fig, axes = plt.subplots(3, figsize=(10, 7), sharex=True)
params = ["H0", "Omega_m", "Omega_l"]
for i in range(3):
    ax = axes[i]
    ax.plot(result_sampling[:, i], "r", alpha=0.5)
    ax.set_ylabel(params[i])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import emcee
import corner
# Datos observacionales de H(z) vs. z
z_obs = np.array([0.07, 0.09, 0.12, 0.17, 0.1791,
                  0.1993, 0.2, 0.24, 0.27, 0.28,
                  0.30, 0.31, 0.34, 0.35, 0.3519,
                  0.36, 0.38, 0.3802, 0.4,
                  0.4004, 0.4247, 0.43, 0.44, 0.4497,
                  0.47, 0.4783, 0.48, 0.51,
                  0.52, 0.56, 0.57, 0.59, 0.5929,
                  0.6, 0.61, 0.64, 0.6797, 0.73, 0.75,
                  0.7812, 0.8754, 0.88, 0.9, 1.037,
                  1.3, 1.363, 1.43, 1.53, 1.75,
                  1.965, 2.3, 2.33, 2.34, 2.36])

H_obs = np.array([69, 69, 68.6, 83, 75,
                  75, 72.9, 79.69, 77, 88.8,
                  81.7, 78.18, 83.8, 82.7,
                  83, 79.94, 81.5, 83, 82.04,
                  77, 87.1, 86.45, 84.81, 92.8,
                  89, 80.9, 87.79, 90.4, 94.35,
                  93.34, 96.8, 98.48, 104, 87.9,
                  97.3, 98.82, 92, 97.3, 98.8, 105,
                  125, 90, 117, 154, 168,
                  160, 177, 140, 202, 186.5,
                  224, 224, 222, 226])

sigma_obs = np.array([19.6, 12, 26.2, 8, 4,
                     5, 29.6, 2.99, 14, 36.64,
                     6.22, 4.74, 3.66, 8.4, 14,
                     3.38, 1.9, 13.5, 2.03, 10.2,
                     11.2, 3.97, 1.83, 12.9, 34,
                     9, 2.03, 1.9, 2.64, 2.3,
                     3.4, 3.18, 13, 6.1, 2.1,
                     2.98, 8, 7, 33.6, 12, 17,
                     40, 23, 20, 17, 33.6,
                     18, 14, 40, 50.4, 8,
                     8, 7, 8])

# Definir el modelo de Hubble
def hubble_model(params, z):
    H0, omega_m, omega_l = params
    omega_k = 1.0 - omega_m - omega_l
    return H0*np.sqrt(omega_m * (1 + z) ** 3 + omega_k * (1 + z) ** 2 + omega_l)

# Definir la log probabilidad
def log_probability(params):
    H0, omega_m, omega_l = params

    # Prior plano
    if not (50 <= H0 <= 100) or not (0 <= omega_m <= 1) or not (0 <= omega_l <= 1):
        return -np.inf

    # Aplicar el modelo
    model = hubble_model(params, z_obs)

    # Calcular el logaritmo del likelihood
    log_like = -0.5 * np.sum(((H_obs - model) / sigma_obs) ** 2)

    return log_like

# Configuración de emcee
nwalkers = 100
ndim = 3
nsteps = 3000

initial = [75, 0.3, 0.7]
pos = initial + 1e-4 * np.random.randn(nwalkers, ndim)

sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability)
sampler.run_mcmc(pos, nsteps, progress=True)
```

```python
#Se imprimen las cadenas
samples = sampler.get_chain(flat=True)
H0_samples, omega_m_samples, omega_l_samples = samples.T

# Tasa de aceptación
print(f'Tasa de aceptación: {np.mean(sampler.acceptance_fraction):.2f}')

# Gráficos
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(H0_samples, color='r', alpha=0.7, label=r'$H_0$')
plt.xlabel('Paso')
plt.ylabel(r'$H_0$')
plt.legend()
plt.subplot(3, 1, 2)
plt.plot(omega_m_samples, color='r', alpha=0.7, label=r'$\Omega_m$')
plt.xlabel('Paso')
plt.ylabel(r'$\Omega_m$')
plt.legend()
plt.subplot(3, 1, 3)
plt.plot(omega_l_samples, color='r', alpha=0.7, label=r'$\Omega_\Lambda$')
plt.xlabel('Paso')
plt.ylabel(r'$\Omega_\Lambda$')
plt.legend()
plt.tight_layout()
plt.show()
```
