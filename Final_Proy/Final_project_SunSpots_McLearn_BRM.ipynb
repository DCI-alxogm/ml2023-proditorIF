El Sol, a diferencia de los planetas rocosos, no posee una superficie sólida, sino que está compuesto principalmente por plasma y presenta varias capas una de estas capas es la fotosfera que es una capa muy delgada y superficial del sol, es en esta capa donde se presentan las manchas solares las cuales se aprecian como regiones más oscuras en la superficie.\
Es posible registrar la energía por unidad de tiempo y área (Flujo) que emite el sol a través de la fotosfera haciendo esto se presentaran zonas donde el flujo registrado sea un tercio en comparación al flujo normal registrado esto es un indicativo de manchas solares.\
Para entender cómo se generan las manchas solares  es fundamental comprender que el Sol posee un campo magnético global generado por la dinamo solar, un proceso en el cual la rotación diferencial del Sol genera corrientes eléctricas en su interior, dando lugar a la generación de campos magnéticos.\
Las manchas solares son regiones donde emerge a la superficie el campo magnético del sol a través de la convección solar, un proceso en el cual el plasma caliente asciende desde las capas internas hasta la fotosfera.
El plasma caliente se mueve en corrientes ascendentes desde el interior estelar hasta la fotosfera , estas corrientes llevan consigo el campo magnético generado en el núcleo, todos estos proceso de convección que  fluyen entre la fotosfera y las zonas internas de la estrella la dan la apariencia granulada a la superficie.\
Como ya se dijo la fotosfera está marcada por una compleja red de campos magnéticos que emergen gracias al flujo del plasma, cuando múltiples líneas de campo magnético emergen en una región específica pueden llegar a entrelazarse y enredarse, creando una estructura magnética compleja que produce una concentración de campos magnéticos en esa área, la formación de manchas solares esta relacionada con estas altas concentraciones   de campo magnético ya que la intensificación del campo magnético local inhibe la convección normal del plasma caliente, impidiendo su ascenso. Esto crea un área relativamente más fría y oscura en comparación con su entorno, lo que percibimos como una mancha solar.\
Cabe destacar que una mancha solar típicamente presenta una umbra oscura en su núcleo, rodeada por una penumbra grisácea con apariencia filamentosa. Estas características proporcionan información valiosa sobre la complejidad y la intensidad del campo magnético en la región de la mancha sola ya que la  presencia de este campo magnético también nos permite explicar por qué el flujo de la umbra era un tercio del registrado en la fotosfera sin manchas. La disminución en el flujo se relaciona con una disminución de la temperatura, a través de la ley de Stefan-Boltzmann.

###**Modelo de dinamo** <a name="Csol"></a>
Para comprender el ciclo solar y su variabilidad, es esencial explorar el modelo propuesto por H.W. Babcock en 1961, que representa uno de los primeros intentos de explicar las propiedades magnéticas observadas en el Sol.

Este modelo, aunque es fenomenológico permite obtener una idea clara de como se comportan los ciclos solares explicándolos mediante cuatro etapas fundamentales:
- **Etapa de Mínimo Solar (A)**: Inicialmente el numero de manchas solares es bajo, en  estos momentos las líneas de campo magnético discurren desde el polo sur al norte. El Sol se comporta como un imán, en el cual los dos hemisferios tienen polaridades opuestas, esto debido a la presencia de un campo dipolo (poloidal) axisimétrico.
- **Etapa de rotación  y estiramiento (B)**:
> - La rotación diferencial del Sol, destacada por Carrington en 1863, causa el estiramiento de las líneas de campo magnético en la dirección este-oeste en el ecuador.
>- Dicho de otra forma la rotación difrencial hace que el campo magnético sumergido se estire en dirección toroidal (envolviendo al Sol). El campo se fortalece con este estiramiento.
- **Etapa de generación de Manchas Solares  (C)**:
> - Las líneas de campo magnético se intensifican y se entrelazan, emergiendo en la superficie mediante los procesos convectivos genrando el par de manchas solares.
> - Cuando las lineas de campo irrumpen en la superficie, crean una región bipolar, emergiendo a través de una de las manchas de cada par y sumergiéndose a través de la otra. Ambas tienen polaridades opuestas. La que circula por delante de su compañera tiene la polaridad de su hemisferio
- **Etapa de cambio de polaridad  (D)**:
>- Las regiones activas, que contienen manchas solares, se desplazan hacia el ecuador.Momento en el cual las manchas conductoras se anulan al tener polaridades inversas.
>- A medida que las manchas conducidas ascienden a los polos, se produce un cambio en la polaridad del Sol, cerrando así el ciclo de 11 años.
>- Tras 11 años, el proceso se reinicia, y para volver a la configuración inicial han de transcurrir 22 años en total, pues las polaridades deben volver a invertirse.

###**Redes nuronales convolucionales (CNN)** <a name="CNN"></a>
Una Red Neuronal Convolucional (CNN) representa un tipo de red neuronal profunda diseñada para procesar datos organizados en una topología en forma de cuadrícula. El término "convolucional" se deriva de la aplicación de la operación matemática de convolución, al menos en una de sus capas.

La convolución, en este contexto, es un operador matemático que se aplica sobre dos funciones f y k, generando una tercera función f⋅k (denotada como g), la cual describe cómo la forma de una función se modifica por la otra. Matemáticamente, esto se expresa como:

$g(i,j)= \sum_{i=-a}^a\sum_{i=-b}^b f(i,j)k(i-a,j-b) $ (EC-b)

Las CNN procesan la entrada de manera secuencial, explorando pequeñas piezas de información y extrayendo detalles sobre diversos patrones. Este proceso se logra mediante un barrido, donde una función denominada filtro o kernel (k) se desplaza sobre toda la entrada, realizando convoluciones en cada posición y generando salidas únicas.

La arquitectura de una red convolucional consta de las siguientes capas, cada una desempeñando un papel crucial en la extracción y transformación de características:

* **Capa de Entrada**: Recibe los datos originales, generalmente representados como una cuadrícula bidimensional, que pueden ser imágenes, por ejemplo.
* **Capa Convolucional**: Aplica la operación de convolución para detectar características específicas en la entrada. Cada neurona en esta capa se enfoca en una región local, permitiendo la identificación de patrones.
* **Capa de Activación**: Introduce funciones de activación no lineales, como la función ReLU (Rectified Linear Unit), para introducir no linealidades en la red y mejorar su capacidad de aprender representaciones complejas.
* **Capa de Agrupación (POOL)**: Reduce la dimensionalidad de la salida de la capa convolucional mediante técnicas como el max pooling o el average pooling. Esto disminuye la carga computacional y proporciona invariancia a pequeñas translaciones en la entrada.
* **Capa FC o Capa Densa**: Convierte la salida de la capa de agrupación en un formato adecuado para la clasificación final. Es una capa completamente conectada que procesa la información a través de todas las neuronas.

Cada una de estas capas desempeña un papel específico en la jerarquía de extracción de características de una CNN, permitiendo el aprendizaje de representaciones complejas y la realización eficiente de tareas como la clasificación de objetos en imágenes.


###**Redes nuronales recurrentes (RNN)** <a name="RNN"></a>
Las redes neuronales recurrentes (RNN) son una clase de aprendizaje profundo basada en los trabajos de David Rumelhart desarrollados en 1986, son un tipo  particular de red neuronal que permite analizar datos de series temporales es decir ahora es posible considerar la dimensión de “tiempo”.\
Es decir, las RNN son un tipo de red neuronal diseñada para trabajar con datos secuenciales, donde la información tiene una estructura temporal. A diferencia de las redes feedforward convencionales, las RNN tienen conexiones que forman ciclos, permitiendo que la información sea reutilizada. Esto genera que las RNN sean particularmente efectivas en la modelación de datos que tienen dependencias temporales, ya que pueden capturar patrones y tendencias en secuencias de eventos a lo largo del tiempo.\
Existen diferentes arquitecturas de redes RNN una de las mas simples son las de tipo unidireccional, esta clase de redes RNN pueden extraer datos de entradas anteriores para hacer predicciones sobre el estado actual.\
En una red (RNN) se incluyen conexiones que “apunta hacia atrás” es decir se implementa una especie de retroalimentación para cada neurona, mas formalmente diríamos que en cada instante de tiempo t  una neurona recurrente recibe como entrada una cierta $x$ que es la salida de la capa anterior , así como su  propia salida del instante de tiempo anterior $y^{t-1}$ para generar su propia salida. Esto implica que ahora cada neurona de la red requiere dos conjuntos de parámetros uno correspondiente a la entrada que viene de la capa anterior y otro correspondiente a la entrada de retroalimentación.\
De forma que la salida para cada nuerona en este caso estada dada por:

>>>>$y_t = f(Wx_t+Uy_{t-1}+b)$ (EC-2)

Donde U representa la matriz de pesos correspondiente a la entrada de retroalimentación la cual también se actualiza durante el entrenamiento de la red.\
Como se puede apreciar ahora se requieren dos matrices de pesos para cada neurona y por lo que se puede decir que las RNN generalmente aumentan la arquitectura de red multicapa convencional con la adición de estos ciclos que conectan nodos adyacentes o pasos de tiempo. \
Ahora dado que la salida de una neurona recurrente en un instante de tiempo t determinado es una función de salidas de instantes de tiempo anteriores lo cual se aprecia clarmanete en (EC-2), se puede decir que una neurona recurrente tiene en cierta forma memoria y precisamente esta memoria es lo que hace altamente eficiente a estas redes en problemas de datos secuenciales.\

Esta representación esta compactada o enrollada, una neurona recurrente se puede desenrollar. El desenrollamiento de una neurona recurrente es un concepto clave para entender cómo las redes neuronales recurrentes (RNN) procesan datos secuenciales a lo largo del tiempo. En lugar de considerar una única instancia de tiempo, el desenrollamiento nos permite visualizar la red a través de múltiples pasos de tiempo.\
Cuando se desenrolla una RNN, se replican las neuronas recurrentes para cada paso de tiempo en la secuencia. Cada replicación se conecta con la siguiente, reflejando la conexión cíclica que caracteriza a las RNN. Esto crea una estructura en forma de cadena, donde cada "copia" de la neurona representa su estado en un instante de tiempo específico.\

Otro concepto crucial en las redes neuronales recurrentes (RNN) es el Backpropagation, que en una red neuronal convencional implica un recorrido inverso para calcular las derivadas parciales del error con respecto a los pesos de las neuronas. Estas derivadas se utilizan en el algoritmo de gradiente descendente para ajustar iterativamente los pesos y minimizar la función de pérdida.

En el contexto de una RNN, el proceso se denomina Backpropagation Through Time (BPTT). Para comprender esto, podemos visualizar la red como una secuencia de redes neuronales, donde se aplica un proceso de Backpropagation que es relativamente equivalente al explicado anteriormente. Sin embargo, en este caso, es necesario incorporar la noción de desenrollamiento, ya que la pérdida en un instante de tiempo t depende del instante anterior.

Al desenrollar la red, extendemos su representación a lo largo de múltiples pasos de tiempo. Durante el BPTT, el error se propaga hacia atrás desde el último hasta el primer instante de tiempo, y esto implica calcular la pérdida para cada momento específico en el desenrollamiento. Este proceso de desenrollamiento es esencial para capturar las dependencias temporales y ajustar los pesos de manera efectiva a lo largo de la secuencia temporal. En resumen, el BPTT combina la retropropagación convencional con la consideración de la estructura secuencial de una RNN para el entrenamiento efectivo de datos secuenciales.





###**Long Short-Term Memory (LSTM)** <a name="LSTM"></a>

Aquí tienes una corrección y una expansión del fragmento:

Las capas de memoria a corto y largo plazo (LSTM) representan una arquitectura de redes neuronales recurrentes (RNN) diseñada específicamente para abordar problemas de dependencia a largo plazo. Pueden considerarse como una extensión de las RNN, mejorando su capacidad para recordar y aprender interacciones pasadas de manera más efectiva.

En contraste con las RNN convencionales, las capas LSTM permiten la retención de información durante periodos más extensos. Cada neurona equipada con LSTM tiene la capacidad única de leer, escribir y borrar información en su memoria. La decisión de almacenar o borrar se toma en función de la importancia de la información de entrada, y esta importancia se pondera mediante pesos ajustables.

Una neurona LSTM consta de tres compuertas esenciales:

* Input Gate (Compuerta de Entrada): Permite la incorporación de nueva información para su almacenamiento.
* Output Gate (Compuerta de Salida): Facilita que la información influya en la salida en el paso de tiempo actual.
* Forget Gate (Compuerta de Olvido): Descarta información no deseada.

Estas compuertas de LSTM son análogas a funciones sigmoide, lo que facilita su inclusión en el proceso de Backpropagation Through Time (BPTT), donde se ajustan los pesos mediante el cálculo de gradientes para optimizar el aprendizaje de la red a lo largo de la secuencia temporal. La capacidad de las capas LSTM para gestionar eficientemente la memoria las convierte en una herramienta poderosa para problemas que implican dependencias temporales a largo plazo.

- **Separación en conjuntos de entrenamiento y prueba**: Los datos se dividen en dos conjuntos principales
-**Generar el formato adecuado para el modelo**: Dependiendo del modelo que se esté utilizando, es posible que se requiera un formato específico de datos.

```python
# Cargar datos
df = pd.read_csv('/content/drive/MyDrive/Machine_Learn/Final_proy/Sunspots.csv')

# Preprocesamiento
sunspots = df['Monthly Mean Total Sunspot Number']
series = np.array(sunspots)
time = np.arange(0, len(series))

# Normalización de datos
scaler = MinMaxScaler()
series_scaled = scaler.fit_transform(series.reshape(-1, 1)).flatten()

# Función para generar datasets de entrenamiento y prueba
def create_sequences(series, window_size):
    X, y = [], []
    for i in range(len(series) - window_size):
        X.append(series[i:i+window_size])
        y.append(series[i+window_size])
    return np.array(X), np.array(y)

# Parámetros
window_size = 64
batch_size = 64
test_size = 0.2

# División de datos en entrenamiento y prueba
train_size = int(len(series_scaled) * (1 - test_size))
train, test = series_scaled[:train_size], series_scaled[train_size:]

# Crear secuencias para entrenamiento y prueba
X_train, y_train = create_sequences(train, window_size)
X_test, y_test = create_sequences(test, window_size)

# Reshape para entrada
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

###**Diseño del modelo**<a name="DModel"></a>
**En esta sección se explica la arquitectura de la red, y el porqué de sus elementos.**

Se eligió emplear un modelo de red neuronal recurrente (RNN) con capas convolucionales y capas LSTM para mejorar la memoria de la red. Aquí hay una breve explicación de cada componente:
- **Conv1D Layer (Convolutional Layer)**:
>- filters=64: Representa el número de filtros convolucionales. Más filtros permiten al modelo aprender patrones más complejos.
>- kernel_size=5: Especifica el tamaño de la ventana del filtro. Un kernel más grande puede capturar patrones temporales más largos.
>-padding='causal': Asegura que la convolución no se realice más allá de la entrada actual, lo cual es crucial para datos secuenciales temporales.
>- Activation='relu': Función de activación rectificadora.

- **LSTM Layers (Long Short-Term Memory)**:

>Dos capas LSTM con 64 unidades y retorno de secuencias. Las capas LSTM son capaces de aprender dependencias temporales a largo plazo y son adecuadas para datos secuenciales.

>La tercera capa LSTM tiene 32 unidades y no retorna secuencias, lo que significa que devuelve solo la última salida de la secuencia temporal.

- **Dense Layers**:

> Dense(16, activation="relu"): Capa densa con 16 unidades y función de activación ReLU. Introduce no linealidad y permite aprender representaciones más complejas.
> Dense(1): Capa de salida con una sola unidad para la predicción de manchas solares.

- **Lambda Layer**:

>Lambda(lambda x: x * 400): Aplica una transformación lineal para ajustar la escala de las predicciones.
En general, este modelo es adecuado para problemas de predicción de series temporales, ya que combina capas convolucionales y recurrentes para capturar patrones tanto locales como temporales a largo plazo.

```python
# Crear modelo
model = Sequential([
    Conv1D(filters=64, kernel_size=5, strides=1, padding='causal', activation='relu', input_shape=[window_size, 1]),
    LSTM(64, return_sequences=True),
    LSTM(64, return_sequences=True),
    LSTM(32),
    Dense(16, activation="relu"),
    Dense(1),
    Lambda(lambda x: x * 400)
])
#Se despliega un grafico de la red neuronal
plot_model(model, show_shapes = True,expand_nested = True,dpi = 80)
```

###**Entrenamiento del modelo**<a name="EntD"></a>
En esta sección se compila y entrena al modelo.

```python
# Compilar modelo
model.compile(loss=tf.keras.losses.Huber(), optimizer='adam', metrics=["mae"])

# Añadir conjunto de validación
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)

# Entrenar modelo
history = model.fit(X_train, y_train, epochs=100, batch_size=batch_size, validation_data=(X_val, y_val), verbose=1)

# Evaluar en conjunto de prueba
test_loss, test_mae = model.evaluate(X_test, y_test, verbose=0)
print(f'Test Loss: {test_loss}, Test MAE: {test_mae}')
```

###**Predicciones y Visualización**<a name="PredV"></a>
En esta sección se realizan las predicciones con el modelo ya entrenado y se grafican los datos iniciales en conjunto con la predicción generada

```python
# Realizar predicciones
predictions = model.predict(X_test)

# Desnormalizar predicciones y datos reales
predictions_denorm = scaler.inverse_transform(predictions)
y_test_denorm = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
# Plotear resultados para conjunto de prueba
plt.figure(figsize=(15, 6))
plt.plot(time[train_size + window_size:], y_test_denorm, label='Actual')
plt.plot(time[train_size + window_size:], predictions_denorm, label='Predicted')
plt.title('Sunspot Number Prediction (Test Set)')
plt.xlabel('Time')
plt.ylabel('Sunspot Number')
plt.legend()
plt.show()
```

```python
# Plotear resultados para conjunto de entrenamiento
predictions_train = model.predict(X_train)
predictions_train_denorm = scaler.inverse_transform(predictions_train)
y_train_denorm = scaler.inverse_transform(y_train.reshape(-1, 1))

plt.figure(figsize=(15, 6))
plt.plot(time[window_size:len(y_train_denorm)+window_size], y_train_denorm, label='Actual')
plt.plot(time[window_size:len(predictions_train_denorm)+window_size], predictions_train_denorm, label='Predicted')
plt.title('Sunspot Number Prediction (Training Set)')
plt.xlabel('Time')
plt.ylabel('Sunspot Number')
plt.legend()
plt.show()
```

```python
# Realizar predicciones en el conjunto completo de datos
full_sequence, full_targets = create_sequences(series_scaled, window_size)
full_sequence = full_sequence.reshape((full_sequence.shape[0], full_sequence.shape[1], 1))
full_predictions = model.predict(full_sequence)

# Desnormalizar predicciones y datos reales para visualización
full_predictions_denorm = scaler.inverse_transform(full_predictions)
full_targets_denorm = scaler.inverse_transform(full_targets.reshape(-1, 1))

# Plotear resultados para el conjunto completo
plt.figure(figsize=(15, 6))
plt.plot(time[window_size:], full_targets_denorm, label='Actual')
plt.plot(time[window_size:], full_predictions_denorm, label='Predicted')
plt.title('Sunspot Number Prediction (Full Dataset)')
plt.xlabel('Time')
plt.ylabel('Sunspot Number')
plt.legend()
plt.show()
```
